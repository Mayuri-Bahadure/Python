## 1) Line Graph
Following Graph shows Each country's gas prices are plotted against the years. 
The plt.figure(figsize=(20, 10)) command sets the figure size to 20x10 inches. 
The plt.title function sets the title of the plot. 
The for loop iterates over each column in the DataFrame (df), excluding the 'Year' column. 
It plots each country's gas prices against the years using plt.plot. 
The marker='.' argument adds a marker to each data point. 
Finally, plt.xticks, plt.xlabel, plt.ylabel, plt.legend, and plt.show are used to customize the plot further and display it.

```python
plt.figure(figsize=(20, 10))#size

plt. title('Gas Prices over Time(in USD)', fontdict={'fontweight':'bold', 'size': 14})

for country in df.columns: #to loop iterates
    if country != 'Year':
        plt.plot(df['Year'], df[country], marker='.', label=country)
        
plt.xticks(df['Year'], fontsize=10)
plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend()
plt.show()
```

## Bar Graph
for creating bar graph calculate average price increase for 8 years (2000-08).

average_increase = df.iloc[10:, 1:].diff().mean(): This line calculates the average annual price increase for each country. It uses .diff() to calculate the difference between consecutive years and .mean() to calculate the average across all years.

plt.figure(figsize=(10, 5)): This line creates a new figure with a specific size (10 inches wide and 5 inches tall) for the plot.

C = ['teal','black','lightblue','pink','purple','lightyellow','grey','magenta','lightgreen','red']: This line defines a list of colors for the bars in the bar chart.

plt.bar(average_increase.index, average_increase, color=C): This line creates a bar chart with each bar representing the average annual price increase for a country. The average_increase.index provides the x-values (countries) and average_increase provides the y-values (average annual price increase).

plt.title('Annual Average Price Increase for Gas prices in USD (2000-8)'): This line sets the title of the plot.

plt.xlabel('Country') and plt.ylabel('Annual Average Price Increase (USD)'): These lines set the labels for the x-axis and y-axis, respectively.

plt.xticks(rotation=45): This line rotates the x-axis labels by 45 degrees for better readability.

plt.grid(axis='y'): This line adds gridlines along the y-axis.

plt.show(): This line displays the plot.

```python
# Calculate average price increase for each country
average_increase = df.iloc[10:, 1:].diff().mean()

plt.figure(figsize=(10, 5))
C = ['teal','black','lightblue','pink','purple','lightyellow','grey','magenta','lightgreen','red']
plt.bar(average_increase.index, average_increase, color=C)
plt.title('Annual Average Price Increase for Gas prices in USD (2000-8)')
plt.xlabel('Country')
plt.ylabel('Anual Avgerage Price Increase (USD)')
plt.xticks(rotation=45)
plt.grid(axis='y')

plt.show()
```

## Pie Chart
The pie chart is created to visualize the distribution of gas prices across different categories or countries. 
gas = df.iloc[:, 1:]: This selects all rows (:) and all columns starting from the second column (index 1). This assumes that the first column contains labels or categories (e.g., country names) and the remaining columns contain the gas prices for each category.

gas_totals = gas.sum(): This calculates the sum of values in each column (category) to get the total gas prices for each category.

plt.figure(figsize=(5, 5)): This creates a new figure for the pie chart with a specified size (5 inches by 5 inches).

plt.pie(gas_totals, labels=gas_totals.index, autopct='%1.1f%%', startangle=140): This creates the pie chart using the pie function. It takes the total gas prices (gas_totals) as the data, the category labels (gas_totals.index) as labels, autopct='%1.1f%%' to format the percentage labels with one decimal place, and startangle=140 to rotate the starting angle of the pie chart.

plt.title('Pie Chart of Gas Prices by Category'): This sets the title of the pie chart.

plt.axis('equal'): This ensures that the pie chart is drawn as a circle by setting the aspect ratio to be equal.

plt.show(): This displays the pie chart.

```python
# Select columns excluding the first one
gas = df.iloc[:, 1:]
# Sum values in each column to get the total for each category
gas_totals = gas.sum()

plt.figure(figsize=(5, 5))
plt.pie(gas_totals, labels=gas_totals.index, autopct='%1.1f%%', startangle=140)

plt.title('Pie Chart of Gas Prices by Category')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```
